```python
import pandas as pd
from collections import Counter
```

```python
# Funktion zum Einlesen der Texte.
def text_vorbereiten(dateiname):
    with open('daten/' + dateiname, 'r') as f:
        lt = f.read()
    return lt
```

```python
df = pd.read_csv('länderübersicht.csv', index_col='index')
```

```python
df['liedtext'] = df['text'].apply(text_vorbereiten)
df
```

```python
df.loc['Vereinigte Staaten']['liedtext']
```

```python
# Welche Wörter kommen in den Hymnen besonder oft vor? 
alle_texte = df['liedtext'].sum().replace('\n', ' ')
word_freqs = Counter(alle_texte.split())
df_wörter = pd.DataFrame.from_dict(word_freqs, orient='index')
```

```python
from nltk.corpus import stopwords
stopWords = set(stopwords.words('german'))

df_wörter[~df_wörter.index.isin(stopWords)].sort_values(by=0, ascending=False)
```
