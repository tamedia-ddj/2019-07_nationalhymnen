```python
# Importieren der nötigen Bibliotheken. 
import pandas as pd
from collections import Counter
from nltk.corpus import stopwords
```

```python
# Funktion zum Einlesen der Texte.
def text_vorbereiten(dateiname):
    with open('daten/' + dateiname, 'r') as f:
        lt = f.read()
    lt = lt.replace('\n', ' ').replace('.', ' ').replace(',', ' ').replace('!', '').replace('?', '').replace(';', '')
    lt = lt.replace('„', '').replace('“', '').replace('"', '')
    lt = lt.replace('  ', ' ')
    return lt
```

```python
# Einlesen der Grunddaten.
df = pd.read_csv('länderübersicht.csv', index_col='index')
```

```python
# Anfügen der Texte der Nationalhymnen. 
df['liedtext'] = df['text'].apply(text_vorbereiten)

# Abspeichern der erweiterten Daten. 
df.to_csv('länderübersicht.csv')
```

```python
# Welche Wörter kommen in den Hymnen besonders oft vor? 
alle_texte = df['liedtext'].sum()
word_freqs = Counter(alle_texte.split())
df_wörter = pd.DataFrame.from_dict(word_freqs, orient='index')

stopWords = list(set(stopwords.words('german')))
stopWords.append('dass')
df_hitparade = df_wörter[~df_wörter.index.str.lower().isin(stopWords)].sort_values(by=0, ascending=False)
df_hitparade = df_hitparade.reset_index()
df_hitparade.head(30)
```
